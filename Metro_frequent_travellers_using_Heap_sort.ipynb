# **Max Heapify Function**

```python
# Define a max heapify function to maintain the max heap property
def max_heapify(arr, n, i):
    largest = i
    left_child = 2 * i + 1
    right_child = 2 * i + 2

    if left_child < n and arr[left_child]['Number_of_Rides'] > arr[largest]['Number_of_Rides']:
        largest = left_child

    if right_child < n and arr[right_child]['Number_of_Rides'] > arr[largest]['Number_of_Rides']:
        largest = right_child

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        max_heapify(arr, n, largest)
```

# **Build Max_heap Function**

```python
# Build a max heap from the DataFrame
def build_max_heap(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        max_heapify(arr, n, i)
```

# **Max-Heap sort Function**

```python
# Sort the travelers using max heap sort
def max_heap_sort(df):
    arr = df.to_dict('records')
    n = len(arr)

    build_max_heap(arr)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        max_heapify(arr, i, 0)

    return pd.DataFrame(arr)
```

# **Grouping Based on Quarters**

```python
grouped=df.groupby(["Quarter"])
```

# **Sorting Quarter 1**

```python
df_Q1=grouped.get_group('Q1')
sorted_df1=max_heap_sort(df_Q1)
```

```python
#print travellers in ascending order
top_travelers_Q1 = sorted_df1.nlargest(5, 'Number_of_Rides')
print("\t\t\t\t\t\tTravellers for Quarter 1")
print("\n\t\t\t\t\t________________________________________")
top_travelers_Q1
```
